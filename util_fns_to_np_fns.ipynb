```python
import os
import re
import numpy as np
from numpy import add
from numpy import subtract as sub
from numpy import multiply as mul
from numpy import divide as div
from numpy import exp
from numpy import sin
from numpy import cos
from numpy import negative as neg
from numpy import reciprocal as inv
from numpy import sqrt
from numpy import log
from numpy import abs
from numpy import power as pow

# Define a mapping from function names to their numpy equivalents
function_mapping = {
    'add': 'np.add',
    'sub': 'np.subtract',
    'mul': 'np.multiply',
    'div': 'np.divide',
    'exp': 'np.exp',
    'sin': 'np.sin',
    'cos': 'np.cos',
    'neg': 'np.negative',
    'inv': 'my_reciprocal',
    'sqrt': 'my_sqrt',
    'log': 'my_log',
    'abs': 'np.abs',
    'pow': 'np.power'
}

# Function to replace function names with their numpy equivalents
def replace_functions(line):
    for func, np_func in function_mapping.items():
        line = line.replace(f'{func}(', f'{np_func}(')
    return line

# Function to replace x0, x1, x2, ... with x[0], x[1], x[2], ...
def replace_variables(line):
    return re.sub(r'\bx(\d+)\b', r'x[\1]', line)

def wrap_with_scaling(new_line, unscaling_factor):
    formula = new_line.split('#')[0]
    ending = new_line.split('#')[1]
    return f"np.multiply({formula}, {unscaling_factor})#{ending}"

# Process each result file
all_results = os.listdir('./results')
for result in all_results:
    if result.endswith('.txt') and not result.endswith('_np.txt') and not 'test' in result:
        with open(f'./results/{result}', 'r') as f:
            lines = f.readlines()
        
        do_scaling = False

        
        # Create a new file with the same name but ending with _np.txt
        new_filename = result.replace('.txt', '_np.txt')
        with open(f'./results/{new_filename}', 'w') as f:
            for i, line in enumerate(lines):
                if "'SCALE'" in line:
                    do_scaling = True
                    unscaling_factor = 1 / float(line.split()[-1])
                    # print(f"Unscaling_factor factor found: {unscaling_factor:.2e}")

                # Check if the line is exactly after '# Resulting function'
                if i > 0 and lines[i-1].strip() == '# Resulting function':
                    # Replace function names with their numpy equivalents
                    new_line = replace_functions(line)

                    # Replace x0, x1, x2, ... with x[0], x[1], x[2], ...
                    new_line = replace_variables(new_line)

                    # Scale if necessary
                    if do_scaling:
                        new_line = wrap_with_scaling(new_line, unscaling_factor)

                    # Add abs inside sqrt and log
                    f.write(new_line)
                else:
                    f.write(line)
                    ...

print("Processing complete.")
```
